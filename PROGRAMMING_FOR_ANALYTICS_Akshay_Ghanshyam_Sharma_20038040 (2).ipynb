Q1 Write a program to develop real-world tax calculation system that generates

Statement of Liability for an employee based on their annual gross income. To

generate the SOL, the system should compute the following:

Universal Social Charge(USC): It is based on income slabs, as per Table 1.1.  

Pay Related Social Insurance(PRSI): It is 4.1% of the gross income.

Income Tax: It is based on tax brackets, mentioned in Table 1.2.

Employees can claim their applicable tax credits from Table 1.3.

Final Tax Liability =  (USC + PRSI + Income Tax) – Total Tax Credits

Net Income = Gross Income – Final Tax Liability. - correct

```python
import math

class Taxsystem:
    """
    Created a class which contains multiple functions to calculate SOL based on income.
    """
    def __init__(self, income, Single, Married, Employee, Rent, Dependents):
        """
        Intialize object for further input of demographic details.

        User input - income (int), Single (str), Married (str), Employee (str), Rent (str), and Dependents (int): Max 2
        """
        self.income = income
        self.Single = Single
        self.Married = Married
        self.Employee = Employee
        self.Rent = Rent
        self.Dependents = Dependents

    def Cal_cred(self):
        """
        Calculates tax laibility based on demographic information.

        Returns: Total tax credits.
        """
        # Performing operation according to Table 1.3: Tax Credits
        credits = 0
        if self.Single.lower() == "yes":
            credits += 2000
        if self.Married.lower() == "yes":
            credits += 4000
        if self.Employee.lower() == "yes":
            credits += 2000
        if self.Rent.lower() == "yes":
            credits += 1000
        if self.Dependents == 0:
            credits += 0
        elif self.Dependents == 1:
            credits += 305
        elif self.Dependents == 2:
            credits += 305 * 2
        elif self.Dependents <= 3:
            credits += 305 * 2
        else:
            print("Invalid Input!")
        return credits

    def Cal_tax(self):
        """
        Calculates income tax based on income brackets.

        Returns:
            float: Income tax amount.
        """
         # Performing operation according to Table 1.2: Income Tax Slabs
        if self.income <= 0:
            print("Invalid Income!")
        elif self.income > 0 and self.income <= 44000:
            tax = self.income * 20 / 100
        else:
            tax = (44000 * 20 / 100) + ((self.income - 44000) * 40 / 100)
        return tax

    def Cal_usc(self):
        """
        Calculates Universal Social Charge based on income.

        Returns: USC amount.
        """
        # Performing operation according to Table 1.1: USC Slabs
        if self.income <= 12012:
            usc = self.income * 0.5 / 100
        elif self.income <= 27382:
            usc = (12012 * 0.5 / 100) + ((self.income - 12012) * 2 / 100)
        elif self.income <= 70044:
            usc = (12012 * 0.5 / 100) + (15370 * 2 / 100) + ((self.income - 27382) * 3 / 100)
        else:
            usc = (12012 * 0.5 / 100) + (15370 * 2 / 100) + (42662 * 3 / 100) + ((self.income - 70044) * 8 / 100)
        return usc

    def prsi(self):
        """
        Calculates Pay Related Social Insurance.

        Returns: PRSI amount.
        """
         # Performing operation according to given PRSI formula
        return math.ceil(self.income * 4.1 / 100)

    def final_tax(self):
        """
        Calculates the final tax liability after applying tax credits.

        Returns: Final tax liability.
        """
        credits = self.Cal_cred()
        income_tax = self.Cal_tax()
        usc = self.Cal_usc()
        prsi = self.prsi()
        total_tax = (income_tax + usc + prsi) - credits #Calculated Final Tax Liability as per the given formula
        return total_tax

    def net_income(self):
        """
        Calculates the net income after deducting the final tax liability.

        Returns: Net income.
        """
        final_tax = self.final_tax()
        return self.income - final_tax  #Calculated net income using the given formula in Net Income Calculation

# User input
employee_name = input("Enter Employee Name: ")
employee_id = input("Enter Employee ID: ")
income = int(input("Enter Annual Gross Income: "))
Single = input("Are you single [Yes/No]? ")
Married = input("Are you Married/Civil Partner [Yes/No]? ")
Employee = input("Are you an employee [Yes/No]? ")
Rent = input("Are you living in rented house [Yes/No]? ")
Dependents = int(input("Number of Dependents [Maximum 2]: "))

# Create a calculator object
calculator = Taxsystem(income, Single, Married, Employee, Rent, Dependents)

# Calculate individual taxes
income_tax = calculator.Cal_tax()
usc = calculator.Cal_usc()
prsi = calculator.prsi()
credits = calculator.Cal_cred()

# Results
final_tax = calculator.final_tax()
net_income = calculator.net_income()

# Output
print("\n**********************************************")
print(f"Statement of liability for: {employee_name}")
print("**********************************************")
print(f"\nGross Income: €{income:.2f}")
print(f"USC: €{usc:.2f}")
print(f"PRSI: €{prsi:.2f}")
print(f"Income Tax: €{income_tax:.2f}")
print("**********************************************")
print(f"\nTotal Tax Credits: €{credits:.2f}")
print(f"Final Tax Liability: €{final_tax:.2f}")
print("**********************************************")
print(f"Net Income: €{net_income:.2f}")
print("**********************************************")
```

Q2 Design and implement an Inventory Management System that allows users to
efficiently manage product stocks. The system should perform the following
operations:
Add a new product: System should not allow users to add the same
product more than once.
Delete a product
Update product details
Lookup a product
On adding, deleting, updating or looking up a product, your program should let the
user know if the operation was successful or not. On looking up a product return the
product name, qty and its price. If the product is not found give the user the
option to add the details they are looking up. The user can perform multiple actions;
they can add a new product and subsequently delete a product without having to
stop and start the program until they decide to quit.

```python
print("***************************************")
print("Welcome to Inventory Management System")
print("***************************************")

# Dictionary to organise product details
Inventory = {} # Stores product info using product IDs as input

def add_product():
    """
    Add new product to inventory.
    """
    product_id = input("Enter the product ID: ")
    if product_id in Inventory:
        print("This product already exists!")
        return

    name = input("Enter the product Name: ")
    try:
        qty = int(input("Enter qty: ")) # Get quantity from the user
        if qty <= 0: # Verify if quantity is postivie
            raise ValueError("qty should be positive.")
        price = float(input("Enter the price per Unit (€): ")) # Get price from the user
        if price <= 0: # Verify if price is positive
            raise ValueError("Price should be positive.")
    except ValueError as e: # Manage invalid input
        print(f"Error: {e}")
        return

    Inventory[product_id] = { # Store product details in dictionary
        "name": name,
        "qty": qty,
        "price": price
    }
    print("Product is added successfully!")

def del_product():
    """
    Delete product from the Inventory.
    """
    product_id = input("Enter Product ID to exclude: ")
    if product_id in Inventory: # Check if product ID already exists
        del Inventory[product_id] # Delete the product from dictionary
        print(f"Product ID {product_id} excluded successfully.")
    else:
        print("Product not found!")

def updt_product():
    """
    Update details of existing product in Inventory.
    """
    product_id = input("Enter Product ID to update: ")
    if product_id not in Inventory: # Check if product ID already exists
        print("Product not found!")
        return

    name = input("Enter new name (Press Enter to skip): ") # Fetch new name (optional)
    try:
        qty = input("Enter new qty (Press Enter to skip): ") # Fetch new quantity (optional)
        qty = int(qty) if qty else None # If input is not empty convert to integer
        price = input("Enter new price (Press Enter to skip): ") # Fetch new price (optional)
        price = float(price) if price else None # If input is not empty convert to float
        if qty is not None and qty <= 0: # Verify if new quantity is positive
            raise ValueError("qty should be positive.")
        if price is not None and price <= 0: # Verify if new price is positive
            raise ValueError("Price should be positive.")
    except ValueError as e: # Manage invalid input
        print(f"Error: {e}")
        return

    if name: # Update name if new name is entered
        Inventory[product_id]["name"] = name
    if qty is not None: # Update quantity if new quantity is entered
        Inventory[product_id]["qty"] = qty
    if price is not None: # Update price if new price is entered
        Inventory[product_id]["price"] = price

    print("Product successfully updated!")

def search_product():
    """
    Search for product in Inventory by ID/name.
    """
    locate_by = input("Search by (id/name): ").strip().lower() # Fetch search option from user
    if locate_by == "id": # Find by product ID
        product_id = input("Enter Product ID: ") # Fetch product details by ID
        product = Inventory.get(product_id)
        if product: # If product is found
            print(f"Product ID: {product_id}")
            print(f"Product Name: {product['name']}")
            print(f"qty: {product['qty']}")
            print(f"Price per unit(€): {product['price']:.2f}")
        else: # If product is not found
            print("Product not found!")
            add_option = input("Do you want to add the product? (yes/no): ").strip().lower()
            if add_option == "yes":
                add_product()
    elif locate_by == "name": # Find by product name
        name = input("Enter Product Name: ").strip().lower()
        found = False
        for product_id, product in Inventory.items(): # Iterate thru all products
            if product["name"].lower() == name: # Verify if product name matches
                print(f"Product ID: {product_id}")
                print(f"Product Name: {product['name']}")
                print(f"qty: {product['qty']}")
                print(f"Price per unit(€): {product['price']:.2f}")
                found = True
        if not found: # If product isn't found
            print("Product not found!")
            add_option = input("Do you want to add the product? (yes/no): ").strip().lower()
            if add_option == "yes":
                add_product()
    else: # Search option invalid
        print("Invalid search option!")

def main():
    """
    Main function to handle user input.
    """
    print("\nAvailable Actions: add, delete, update, search, quit")
    while True:
        action = input("\nEnter action: ").strip().lower() # Fetch action from the user
        if action == "add":
            add_product()
        elif action == "delete":
            del_product()
        elif action == "update":
            updt_product()
        elif action == "search":
            search_product()
        elif action == "quit":
            print("Quitting the Inventory Management System.")
            break
        else:
            print("Invalid action! Please choose valid option.")

# Start the program
main()
```

Q4 Write a program to reverse the string without reversing its individual words.
Words are separated by spaces. The string may contain leading or trailing spaces,
as well as multiple spaces between words. The program should return string that
only have a single space separating the words, and no extra spaces. Ensure that
punctuation marks are kept in the correct positions after reversing the words. It
should also result the length of each word along with the word itself in the
reversed order. If the string is empty, return an empty string.
Sample:
Input: “  I  am reversing this string.  “
Output:  
Length of each word: [(‘string.’, 7), (‘this’, 4), (‘reversing’, 9), (‘am’, 2), (‘I’, 1)]
Reversed String: “string. this reversing am I”

```python
def reverse_string_with_word_lengths(text):
    """
    Reverses a string without reversing its individual words.
    Returns: A tuple containing:
            - A list of tuples, in which each tuple includes a word and its length.
            - The reversed string along with single spaces and no leading or trailing spaces.
    """
    if not text:
        return [], ""

    # exclude leading/trailing spaces and split the string into words
    words = text.strip().split()

    # Filter out empty strings that might result from multiple spaces
    words = [word for word in words if word]

    # Reverse the list of words
    reversed_words = words[::-1]

    # Create the list of (word, length) tuples
    word_lengths = [(word, len(word)) for word in reversed_words]

    # Join the reversed words with single spaces
    reversed_string = " ".join(reversed_words)

    return word_lengths, reversed_string

# Sample usage:
input_string = "  I am reversing this string.  "
word_lengths, reversed_string = reverse_string_with_word_lengths(input_string)
print("Length of each word:", word_lengths)
print("Reversed String:", reversed_string)
```
